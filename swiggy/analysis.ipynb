```python
import pandas as pd

# 1. Load dataset
df = pd.read_csv("Swiggy_Order_Dataset (1).csv")

# 2. View dataset info
print(df.info())
print(df.head())

# 3. Remove duplicate rows
df = df.drop_duplicates()

# 4. Check missing values
print(df.isnull().sum())

# 5. Convert Order_Date_Time to datetime
df['Order_Date_Time'] = pd.to_datetime(df['Order_Date_Time'], errors='coerce')

# 6. Create new time features
df['Year'] = df['Order_Date_Time'].dt.year
df['Month'] = df['Order_Date_Time'].dt.month
df['Day'] = df['Order_Date_Time'].dt.day
df['Hour'] = df['Order_Date_Time'].dt.hour

# 7. Standardize text columns (remove extra spaces)
df['City '] = df['City '].str.strip()
df['Cuisine_Type'] = df['Cuisine_Type'].str.strip()
df['Payment_Method'] = df['Payment_Method'].str.strip()

# 8. Check for outliers or incorrect values
print(df.describe())

# 9. Rename columns (optional for better readability)
df.columns = df.columns.str.replace(" ", "_")

# 10. Save cleaned dataset
df.to_csv("Swiggy_Cleaned_Dataset.csv", index=False)

print("Data cleaning completed and file saved.")
```
